```python
from selenium.webdriver.chrome.options import Options
options = Options()
options.add_argument("--disable-notifications")
# options.add_argument('--headless')
# options.add_argument('user-agent="Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36"')
    
def get_description(url):   
    time.sleep(2 + random.randint(1,10)*0.1)
    driver = webdriver.Chrome(options = options)
    driver.find_element_by_tag_name('body').click()
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'lxml')

    text_list = []

    first = soup.find('span',{'itemprop':'Description'})
    middle = soup.find('span',{'id':'product-read-more-text-mq1'})
    last = soup.find('span',{'id':'product-read-more-text'})

    try:
        review_count = soup.find('a', {'href':'#product-reviews-section'})
        review_count = int(review_count.text.split(' ')[1])
    except:
        review_count = np.nan
    
    try:
        rating = soup.find('div',{'class':'pr-snippet-rating-decimal'})
        rating = float(rating.text)
    except:
        rating = np.nan
        
    driver.close()
    print('souping')
    if first:
        first = first.text.strip()
        text_list.append(first)

    if middle:
        middle = middle.text.strip()
        text_list.append(middle)

    if last:
        last = last.text.strip()
        text_list.append(last)

    description = ''
    
    for text in text_list:
        description += text
 
    return description, rating, review_count
```
